```python
import pandas as pd
import numpy as np
data = pd.read_excel('Python_time_test.xlsx')
data.reset_index(drop=True, inplace=True)
data['Order_time'] = pd.to_datetime(data['Order_time'])
data.drop_duplicates(inplace=True)
data.sort_values(by=['User', 'Order_time'], ascending=[True, True])
data
```

```python
res = pd.DataFrame([], columns=['User', 'MeanGap', 'MaxGap', 'MinGap'])

users = sorted(list(set(data.User)))
for u in users:
    temp = data[data['User']==u]
    gaps = []
    if len(temp)==1:# 若只有一次记录，时间差计为0
        gaps = [0]
    else:
        for i in range(temp.shape[0]-1):
            gaps.append((temp.iloc[i+1,:]['Order_time'] - temp.iloc[i,:]['Order_time']).days)
    
    res.loc[res.shape[0]+1] = {'User':u, 
                               'MeanGap':round(np.mean(gaps),2), 
                               'MaxGap':round(np.max(gaps),2), 
                               'MinGap':round(np.min(gaps),2)}
        
res
```
